# 3. 데이터 시각화(비디오 서버)

```python
video = joined[joined['시설물종류명'] == 'VES']
```

```python
video.head()
```

```python
df_count = pd.DataFrame(video['dong'].value_counts())
df_count
```

```python
def mapping_dong_video(x):
    if x['dong'] in df_count['dong']:
        return df_count.loc[x['dong']]['dong']
    else:
        return 0
```

```python
gdf_video = gdf.copy()
gdf_video['val'] = gdf_video.apply(mapping_dong_video, axis=1)
gdf_video
```

```python
import folium
from folium.plugins import HeatMap
import geopandas as gpd

# GeoDataFrame 생성
gdf_video = gpd.GeoDataFrame(gdf_video, geometry='geometry')

# 중심 좌표 설정
# 지도 중심점 설정
center = [37.5665, 126.9784]

# folium 지도 생성
m = folium.Map(location=center, zoom_start=11)

# Choropleth 클래스를 사용하여 지도에 시각화
folium.Choropleth(
    geo_data=gdf_video,
    name='choropleth',
    data=gdf_video,
    columns=['gid', 'val'],
    key_on='feature.properties.gid',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Value'
).add_to(m)

# 지도에 행정동명 추가
folium.GeoJson(
    gdf_video,
    name='geojson',
    tooltip=folium.features.GeoJsonTooltip(fields=['dong'], aliases=['동'], localize=True),
).add_to(m)

# 지도를 HTML 파일로 저장
m.save('map.html')

m
```

# 4. 인구밀도 대비 데이터 시각화

```python
# 인구밀도 대비
gdf_video_people = gdf.copy()
```

```python
df_count = pd.DataFrame(video['dong'].value_counts())
df_count
```

```python
gdf_dong_index = gdf.set_index('dong')
gdf_dong_index
```

```python
def mapping_dong_video_people(x):
    if x['dong'] in df_count['dong']:
        return gdf_dong_index.loc[x['dong']]['val'] // df_count.loc[x['dong']]['dong']
    else:
        return 0
```

```python
gdf_video_people['val'] = gdf_video_people.apply(mapping_dong_video_people, axis=1)
gdf_video_people
```

```python
import folium
from folium.plugins import HeatMap
import geopandas as gpd

# GeoDataFrame 생성
gdf_video_people = gpd.GeoDataFrame(gdf_video_people, geometry='geometry')

# 중심 좌표 설정
# 지도 중심점 설정
center = [37.5665, 126.9784]

# folium 지도 생성
m = folium.Map(location=center, zoom_start=11)

# Choropleth 클래스를 사용하여 지도에 시각화
folium.Choropleth(
    geo_data=gdf_video_people,
    name='choropleth',
    data=gdf_video_people,
    columns=['gid', 'val'],
    key_on='feature.properties.gid',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Value'
).add_to(m)

# 지도에 행정동명 추가
folium.GeoJson(
    gdf_video_people,
    name='geojson',
    tooltip=folium.features.GeoJsonTooltip(fields=['dong'], aliases=['동'], localize=True),
).add_to(m)

# 지도를 HTML 파일로 저장
m.save('map.html')

m
```

# 5. 스코어링
- 인구밀도 대비 적을수록 높은 값을 주도록

```python
score_video_people = pd.DataFrame(gdf_video_people)[['dong', 'gid', 'val']]
score_video_people['val'] = score_video_people['val'].apply(lambda x: score_video_people['val'].max()+1 if x == 0 else x)
score_video_people = score_video_people.sort_values(by='val', ascending=False).reset_index().drop(['index'], axis=1).reset_index()
score_video_people
```
